```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
salario = np.array([100,200,300,500,400,150])
salario
```

```python
plt.plot(salario)
```

```python
plt.plot(salario, c='black')
```

```python
plt.plot(salario, c='black', ls='--')
```

```python
plt.plot(salario, c='black', ls='--', marker='s')
```

```python
plt.plot(salario, c='red', ls='--', marker='s', ms=13)
```

```python
plt.plot(salario, c='red', ls='--', marker='s')
plt.show()
```

```python
salarios_marcos = np.array([100,200,400,300])
salarios_gileno = np.array([50,300,500,450])
```

```python
# add label
plt.plot(salarios_marcos, label='Marcos')
plt.plot(salarios_gileno, c='red', label='Gileno')
plt.show()
```

```python
# add legenda
plt.plot(salarios_marcos, marker='s', label='Marcos')
plt.plot(salarios_gileno, c='red', marker='^', label='Gileno')
plt.legend()
plt.show()
```

```python
# add legenda
plt.plot(salarios_marcos, marker='s', label='Marcos')
plt.plot(salarios_gileno, c='red', marker='^', label='Gileno')
plt.legend(loc='lower center')
plt.show()
```

```python
# add linha
salario_pedro = [200,150,300,700]
plt.plot(salarios_marcos, ls='--', marker='s', label='Marcos')
plt.plot(salarios_gileno, ls='--', c='red', marker='^', label='Gileno')
plt.plot(salario_pedro, ls='--', c='green', marker='o', label='Pedro')
plt.legend(loc='upper center')
plt.show()
```
